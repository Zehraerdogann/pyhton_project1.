```python
df_group=df.groupby(['City','Churn Reason'])['Total Charges'].agg(['mean']).dropna()
```

```python


top_reasons_counts = df['Churn Reason'].value_counts().head(10)

plt.figure(figsize=(7, 5))
sns.barplot(x=top_reasons_counts.values, y=top_reasons_counts.index, palette='viridis')
plt.title('Top 10 Churn Reasons')
plt.xlabel('Count')
plt.ylabel('Churn Reason')
plt.show()
```

```python
df.groupby(['Gender'])['Total Revenue'].sum().dropna()
```

```python

gender_revenue = df.groupby('Gender')['Total Revenue'].sum()


plt.figure(figsize=(6, 6))
plt.pie(gender_revenue, labels=gender_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Revenue According to Gender')
plt.show()
```

```python

plt.figure(figsize=(7, 5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Generally, the age distribution shows that the number of young individuals is higher compared to other age groups, but the number of individuals decreases as age advances.


```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Churn Category', y='Total Revenue', data=df)
plt.title('Total Revenue by Churn Category')
plt.xlabel('Churn Category')
plt.ylabel('Total Revenue')
plt.show()
```

The "Competitor" category shows the widest range of total income, while the other categories have different income distributions as well. The "Price" category appears to have a lower central tendency compared to the others. The outliers in the "Dissatisfaction" and "Competitor" categories indicate that there are some individuals in these groups who earn high incomes. The chart allows for examining the differences in total revenues and potential income losses based on churn categories.

```python
# Filtering numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

#Calculating the correlation matrix
correlation_matrix = numeric_df.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Correlations:

Age and Many Variables: A negative relationship is observed between age and some variables. Notably, the correlation of age with "Total Income" and "Total Salaries" is remarkable.
Monthly Salary and Total Income: There is a strong positive relationship between "Monthly Salary" and "Total Income."
Total Salaries and Other Variables: There are also positive relationships between "Total Salaries" and some other variables.
General Assessment:
The graph provides a visual representation of the relationships between variables, helping to understand which variables are related to each other during data analysis. Positive and negative correlations provide important insights into how specific variables affect one another.
